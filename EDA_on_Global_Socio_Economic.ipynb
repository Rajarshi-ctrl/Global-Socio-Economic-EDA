## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data Wrangling - Preparing the dataset for analysis

# Step 1: Drop the unnamed last column (all NaN)
df_data_clean = df_data.drop(columns=['Unnamed: 69'], errors='ignore')
print('Step 1: Dropped empty Unnamed:69 column')
print('Shape after drop:', df_data_clean.shape)

# Step 2: Melt wide format to long format for easier analysis
year_cols = [str(y) for y in range(1970, 2016) if str(y) in df_data_clean.columns]
id_cols = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']

df_long = df_data_clean.melt(
    id_vars=id_cols,
    value_vars=year_cols,
    var_name='Year',
    value_name='Value'
)
df_long['Year'] = df_long['Year'].astype(int)
df_long = df_long.dropna(subset=['Value'])
print('Step 2: Melted to long format - removed NaN rows')
print('Long format shape:', df_long.shape)

# Step 3: Filter key indicators for analysis
key_indicators = [
    'Literacy rate, adult total (% of people ages 15 and above)',
    'Government expenditure on education, total (% of GDP)',
    'School enrollment, primary (% gross)',
    'School enrollment, secondary (% gross)',
    'School enrollment, tertiary (% gross)',
    'Population, total'
]
df_key = df_long[df_long['Indicator Name'].isin(key_indicators)]
print('Step 3: Filtered key indicators')
print('Key indicators shape:', df_key.shape)

# Step 4: Merge with country metadata
df_merged = df_long.merge(
    df_country[['Country Code', 'Region', 'Income Group']],
    on='Country Code',
    how='left'
)
df_merged = df_merged.dropna(subset=['Region'])
print('Step 4: Merged with country metadata (Region, Income Group)')
print('Merged dataset shape:', df_merged.shape)

print()
print('Data Wrangling Complete! Datasets ready for visualization.')
````

### What all manipulations have you done and insights you found?

Data Wrangling Steps and Insights: (1) Dropped 'Unnamed: 69' column (100% NaN, useless column). (2) Converted wide format (886K rows x 70 cols) to long format - resulting in 4.19M rows x 6 cols; removed all NaN value rows, retaining only actual measurements. (3) Extracted 6 key education indicators (literacy, enrollment at primary/secondary/tertiary levels, government expenditure, population) for focused analysis. (4) Merged main data with EdStatsCountry metadata to add Region and Income Group dimensions - resulting in 3.9M rows with geographic context. Key insight: Most data is concentrated in 1990-2015 period. Developing regions show significantly higher missing rates than developed regions.

#### Chart - 4

````python
# Chart - 4: Bivariate - Line Chart (Numerical-Numerical)
# Trend of Average Education Indicator Value over Years by Region

# Filter relevant region column and year columns
year_cols = [c for c in df_data.columns if c.isdigit() and 1990 <= int(c) <= 2020]
df_time = df_data[['Country Code'] + year_cols].copy()

# Merge with country data to get region
df_merged_region = df_time.merge(df_country[['Country Code', 'Region']].dropna(), on='Country Code', how='inner')
df_merged_region = df_merged_region[df_merged_region['Region'] != '']

# Melt to long format
df_region_long = df_merged_region.melt(id_vars=['Country Code', 'Region'], value_vars=year_cols, var_name='Year', value_name='Value')
df_region_long['Year'] = df_region_long['Year'].astype(int)
df_region_long['Value'] = pd.to_numeric(df_region_long['Value'], errors='coerce')

# Group by Region and Year
df_region_trend = df_region_long.groupby(['Region', 'Year'])['Value'].mean().reset_index()

fig, ax = plt.subplots(figsize=(13, 6))
for region in df_region_trend['Region'].unique():
    subset = df_region_trend[df_region_trend['Region'] == region]
    ax.plot(subset['Year'], subset['Value'], marker='o', markersize=3, label=region)

ax.set_title('Average Education Indicator Trend by Region (1990-2020) - Chart 4', fontsize=13, fontweight='bold')
ax.set_xlabel('Year', fontsize=11)
ax.set_ylabel('Average Indicator Value', fontsize=11)
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
````

````python
# Chart - 11: Grouped Bar Chart - Primary vs Secondary Enrollment Rate by Region
# Chart Type: Bivariate (Categorical - Numerical)
# Purpose: Compare primary and secondary school enrollment rates across world regions

import warnings
warnings.filterwarnings('ignore')

# Indicator codes
primary_enroll_code = 'SE.PRM.ENRR'    # School enrollment, primary (% gross)
secondary_enroll_code = 'SE.SEC.ENRR'  # School enrollment, secondary (% gross)

year_cols_list = [c for c in df_data.columns if c.isdigit()]

def get_mean_by_region(indicator_code):
    subset = df_data[df_data['Indicator Code'] == indicator_code][['Country Code'] + year_cols_list].copy()
    subset['mean_val'] = subset[year_cols_list].mean(axis=1, skipna=True)
    merged = subset[['Country Code', 'mean_val']].merge(
        df_country[['Country Code', 'Region']], on='Country Code'
    ).dropna()
    merged = merged[merged['Region'].str.strip() != '']
    return merged.groupby('Region')['mean_val'].mean()

primary_data = get_mean_by_region(primary_enroll_code)
secondary_data = get_mean_by_region(secondary_enroll_code)

df_grouped = pd.DataFrame({'Primary Enrollment': primary_data,
                           'Secondary Enrollment': secondary_data}).dropna()
print(f"Regions: {df_grouped.shape[0]}")

# --- Plot ---
fig, ax = plt.subplots(figsize=(13, 7))
x = range(len(df_grouped))
width = 0.35
ax.bar([i - width/2 for i in x], df_grouped['Primary Enrollment'],
       width=width, label='Primary Enrollment (%)', color='#1976D2', alpha=0.85)
ax.bar([i + width/2 for i in x], df_grouped['Secondary Enrollment'],
       width=width, label='Secondary Enrollment (%)', color='#F57C00', alpha=0.85)
ax.set_xticks(list(x))
ax.set_xticklabels(df_grouped.index, rotation=25, ha='right')
ax.set_ylabel('Average Enrollment Rate (% Gross)', fontsize=11)
ax.set_title('Chart 11: Primary vs Secondary School Enrollment Rate by Region',
             fontsize=12, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
print("Chart 11 rendered successfully.")
````

##### 1. Why did you pick the specific chart?

A **Grouped Bar Chart** is the ideal choice for this **Bivariate (Categorical – Numerical)** analysis because it enables direct side-by-side comparison of two related numerical metrics — primary and secondary enrollment rates — across multiple regional categories simultaneously. Unlike a stacked bar chart that shows composition, a grouped bar chart preserves the absolute magnitudes of each metric, making it easy to see both the difference between primary and secondary enrollment within each region, and how each enrollment type varies across regions.

````python
# Chart - 12: Area Chart - Trend of Female vs Male Tertiary Enrollment Over Years
# Chart Type: Bivariate (Numerical - Numerical / Time-Series)
# Purpose: Track and compare gender trends in tertiary (university-level) education enrollment

import warnings
warnings.filterwarnings('ignore')

# Indicator codes for tertiary enrollment by gender
female_tertiary_code = 'SE.TER.ENRR.FE'  # Tertiary enrollment, female (% gross)
male_tertiary_code = 'SE.TER.ENRR.MA'    # Tertiary enrollment, male (% gross)

year_cols_list = [c for c in df_data.columns if c.isdigit()]

def get_global_avg_by_year(indicator_code):
    """Compute global average of an indicator for each year."""
    subset = df_data[df_data['Indicator Code'] == indicator_code][year_cols_list].copy()
    return subset.mean(axis=0, skipna=True)  # Series: year -> average

female_avg = get_global_avg_by_year(female_tertiary_code)
male_avg = get_global_avg_by_year(male_tertiary_code)

# Convert to DataFrame
df_gender_trend = pd.DataFrame({'Female': female_avg, 'Male': male_avg})
df_gender_trend.index = df_gender_trend.index.astype(int)
df_gender_trend = df_gender_trend[(df_gender_trend.index >= 1970) & (df_gender_trend.index <= 2020)]
df_gender_trend = df_gender_trend.dropna(how='all')

print(f"Years with data: {df_gender_trend.shape[0]}")

# --- Plot ---
fig, ax = plt.subplots(figsize=(13, 6))
ax.fill_between(df_gender_trend.index, df_gender_trend['Female'],
                alpha=0.5, color='#E91E63', label='Female Tertiary Enrollment (%)')
ax.fill_between(df_gender_trend.index, df_gender_trend['Male'],
                alpha=0.5, color='#1565C0', label='Male Tertiary Enrollment (%)')
ax.plot(df_gender_trend.index, df_gender_trend['Female'], color='#C2185B', linewidth=1.5)
ax.plot(df_gender_trend.index, df_gender_trend['Male'], color='#0D47A1', linewidth=1.5)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Average Tertiary Enrollment Rate (% Gross)', fontsize=11)
ax.set_title('Chart 12: Global Female vs Male Tertiary Enrollment Trend (Area Chart)',
             fontsize=12, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
print("Chart 12 rendered successfully.")
````

The filled areas allow visual comparison of the magnitude of each gender's enrollment across decades, while the overlap region (if any) highlights periods of convergence or divergence.
